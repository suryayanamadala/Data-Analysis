```python
import pandas as pd
import json
```

```python
# Loading the JSON data
with open('data_set/match_summary.json', 'r') as file:
    json_data = json.load(file)

# Creating a DataFrame from the match summary of the first record
m_summary = pd.DataFrame(json_data[0]['matchSummary'])

# Displaying the first ten records
m_summary.head(10)
```

```python
m_summary.shape
```

```python
m_summary = m_summary.rename(columns={'scorecard': 'match_id'})
m_summary.head()
```

```python
match_ids_dict = {}

for _, row in m_summary.iterrows():
    key1 = f"{row['team1']} Vs {row['team2']}"
    key2 = f"{row['team2']} Vs {row['team1']}"
    match_id = row['match_id']
    match_ids_dict[key1] = match_id
    match_ids_dict[key2] = match_id
```

```python
import os

# Checking if the directory exists, and create it if it does not
directory = 'csv_set'
if not os.path.exists(directory):
    os.makedirs(directory)

# Now saving the DataFrame to the CSV file
m_summary.to_csv(f'{directory}/dim_match_summary.csv', index=False)
```

```python
# Opening and reading data from the JSON file
with open('data_set/batting_summary.json') as f:
    data = json.load(f)  # Loading JSON data into a Python object
    all_records = []  # Initializing an empty list to hold all records

    # Iterating through each record in the JSON data
    for rec in data:
        # Extending the all_records list with the contents of 'battingSummary'
        all_records.extend(rec['battingSummary'])
  
# Creating a DataFrame from the collected batting summaries
m_batting = pd.DataFrame(all_records)

# Displaying the first 10 rows of the DataFrame for a quick preview
m_batting.head(10)
```

```python
# Adding the 'out/not_out' column based on the 'dismissal' column
m_batting['out/not_out'] = m_batting['dismissal'].apply(lambda x: 'out' if x else 'not_out')

# Displaying the first 10 rows of the DataFrame
m_batting.head(10)
```

```python

# Adding the 'match_id' column by mapping 'match' column values using match_ids_dict
m_batting['match_id'] = m_batting['match'].map(match_ids_dict)

# Displaying the first few rows of the DataFrame to verify the new column
m_batting.head()
```

```python
# Removing the 'dismissal' column from the df_batting DataFrame
m_batting.drop(columns=["dismissal"], inplace=True)

# Displaying the first 10 rows of the updated DataFrame
m_batting.head(10)
```

```python
m_batting['batsmanName'] = m_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
m_batting['batsmanName'] = m_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
m_batting.head(10)
```

```python
m_batting.shape
```

```python
m_batting.to_csv('csv_set/fact_bating_summary.csv', index = False)
```
